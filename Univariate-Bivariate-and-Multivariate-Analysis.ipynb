```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
df.head()
```

```python
df.shape
```

### Univariate Analysis

```python
df.species.unique()
```

```python
df_setosa = df.loc[df.species=='setosa']
```

```python
df_virginica = df.loc[df.species=='virginica']
```

```python
df_versicolor = df.loc[df.species=='versicolor']
```

```python
df_versicolor.head()
```

```python
df_virginica.head()
```

```python
plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']), 'o')
plt.plot(df_virginica['sepal_length'], np.zeros_like(df_virginica['sepal_length']), 'o')
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']), 'o')
plt.xlabel('Sepal Length')
plt.show()
```

### Bivariate Analysis

```python
sns.FacetGrid(df, hue="species", size=5).map(plt.scatter,"petal_length", "sepal_width").add_legend()
# here, hue is target value i.e., species here
```

#### From the above bivariate analysis via Scatter plot, we can easily determine that-  if we consider petal_length and sepal_width along x and y-axis, then, setosa can be easily classified by plotting a line between it and the other 2 species. But, some of the points of versicolor and virginica are overlapping on each other. Thus, we can't be able to classify them, by just drawing a straight line between them.

### Multivariate Analysis

```python
sns.pairplot(df, hue="species", size=3)
```

#### i.) If we have n=more than 3 features, then, we use pairplot to understand how those features are inter-related, and this is being compared with each and every feature.This is the most basic thing in Exploratory Data Analysis.

#### ii.) We can see which are the most important features, which will help to classify all the points easily.

#### iii.) How these features are correlated i.e., Pearson Correlation? e.g., in the above plot, if it's a regression problem for petal_length(y) v/s petal_width(x), then, it is a positive correlation.
